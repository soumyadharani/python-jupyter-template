### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.

import matplotlib.pyplot as plt
import numpy as np

# Data for the plots
x = np.linspace(0, 10, 100)
y = x**2

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
pie_sizes = [40, 25, 20, 15]
pie_colors = ['blue', 'red', 'green', 'purple']

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Line chart
axs[0, 0].plot(x, y, label=r'$y = x^2$', color='b', linestyle='-', linewidth=2)
axs[0, 0].set_title("Line Plot of y = x^2", fontsize=14)
axs[0, 0].set_xlabel("X Values", fontsize=12)
axs[0, 0].set_ylabel("Y Values", fontsize=12)
axs[0, 0].grid(True, linestyle='--', alpha=0.6)
axs[0, 0].legend()

# Bar chart
axs[0, 1].bar(categories, values, color=pie_colors)
axs[0, 1].set_title("Bar Plot of Categories", fontsize=14)
axs[0, 1].set_xlabel("Categories", fontsize=12)
axs[0, 1].set_ylabel("Values", fontsize=12)
axs[0, 1].grid(axis='y', linestyle='--', alpha=0.6)

# Scatter plot
axs[1, 0].scatter(x_scatter, y_scatter, color='blue', marker='o', s=100, edgecolors='black')
for i in range(len(x_scatter)):
    axs[1, 0].text(x_scatter[i], y_scatter[i] + 0.3, f'({x_scatter[i]}, {y_scatter[i]})', ha='center', fontsize=10)
axs[1, 0].set_title("Scatter Plot with Annotations", fontsize=14)
axs[1, 0].set_xlabel("X Values", fontsize=12)
axs[1, 0].set_ylabel("Y Values", fontsize=12)
axs[1, 0].grid(True, linestyle='--', alpha=0.6)

# Pie chart
axs[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', colors=pie_colors, startangle=140, wedgeprops={'edgecolor': 'black'})
axs[1, 1].set_title("Programming Language Popularity", fontsize=14)

# Adjust layout
plt.tight_layout()
plt.show()
```
