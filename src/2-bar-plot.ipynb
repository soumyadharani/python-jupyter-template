### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
# Import necessary library
import matplotlib.pyplot as plt

# Define data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'purple']

# Create figure and axis
plt.figure(figsize=(6, 4))
plt.bar(categories, values, color=colors, width=0.6)

# Add labels and title
plt.title("Category-wise Bar Plot", fontsize=14)
plt.xlabel("Categories", fontsize=12)
plt.ylabel("Values", fontsize=12)

# Add grid lines
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Display values on top of bars
for i, v in enumerate(values):
    plt.text(i, v + 0.3, str(v), ha='center', fontsize=12)

# Show the plot
plt.show()


```
